### Complaint Trends Over Time
Next, I examine how complaint volume changes over time. This helps determine whether telecommunications complaints increase or decrease across different years.

```python
cleaned["Year"] = cleaned["Date Created"].dt.year
year_counts = cleaned["Year"].value_counts().sort_index()
year_counts
```

```python
year_counts.plot(kind="line", marker="o")
plt.title("Complaints Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Complaints")
plt.show()
```

The number of complaints fluctuates across years. Changes in complaint volume may reflect regulatory changes, shifts in telecommunications technology, or increased awareness of consumer complaint reporting systems.

### Geographic Distribution of Complaints
Another aspect of the analysis is examining complaints by state. This helps identify whether certain regions generate more telecommunications complaints.

```python
state_counts = cleaned["State"].value_counts()
state_counts.head(10)
```

```python
state_counts.head(10).plot(kind="bar")
plt.title("Top States by Complaint Volume")
plt.xlabel("State")
plt.ylabel("Number of Complaints")
plt.show()
```

Larger states such as California and Texas appear among the states with the highest complaint counts. This likely reflects population size rather than necessarily indicating worse telecommunications services.

### Relationship Between Complaint Type and State
To better understand how complaint types vary across regions, I create a heatmap that compares complaint categories across the states with the highest complaint volume.

```python
import seaborn as sns

state_issue = pd.crosstab(cleaned["State"], cleaned["Issue"])

top_states = state_issue.sum(axis=1).sort_values(ascending=False).head(10).index
heatmap_data = state_issue.loc[top_states]

plt.figure(figsize=(12,6))
sns.heatmap(heatmap_data, cmap="Blues")
plt.title("Complaint Types by State")
plt.show()
```

The heatmap visualizes the relationship between U.S. states and the types of complaints reported in the dataset. Each cell represents the frequency of a specific complaint category within a particular state. Darker colors indicate higher concentrations of complaints.

The heatmap shows that Unwanted Calls appear consistently across many states, suggesting that this issue is widespread across the country rather than limited to a specific region. While some states appear to report slightly more complaints than others, the overall pattern indicates that telecommunications related issues are experienced nationwide.

This visualization helps reveal geographic patterns in the data and reinforces the earlier finding that unwanted calls are the most common complaint category in the dataset.

### Complaint Trends Over Time

To better understand how telecommunications complaints change over time, I created a line chart showing the number of complaints recorded by date. This visualization helps identify trends in complaint activity and highlights whether complaint volumes increase or decrease during certain periods.

By examining complaint activity over time, we can gain insight into whether consumer issues remain stable or fluctuate throughout the dataset. This type of visualization can also help identify periods where complaints spike, which may indicate changes in telecommunications practices or increased consumer awareness.

```python
complaints_over_time = cleaned.groupby("Date Created").size()

plt.figure(figsize=(12,6))
complaints_over_time.plot()
plt.title("Telecommunications Complaints Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Complaints")
plt.show()
```

### Analysis of Complaint Trends

The line chart illustrates how telecommunications complaints change over time based on the date they were created. By grouping complaints by date, the visualization helps reveal patterns in consumer reporting behavior.

The trend shows how complaint activity fluctuates over time, with some periods experiencing higher numbers of complaints than others. These changes may reflect shifts in telecommunications practices, seasonal patterns in consumer behavior, or periods when robocalls and other service issues become more prevalent.

Visualizing complaints over time provides useful context for understanding how consumer concerns evolve and helps highlight periods where complaint activity increases or decreases.
